```python
def visual_majority_count(data):
    plt.figure(figsize=(12,6))
    g = sb.countplot(data.Sex)
    plt.title('MAJORITY PASSENGER COUNTS', weight='bold')
    plt.grid(linestyle = '--', axis='y')
    
    
    
visual_majority_count(A)
```

```python
def visual_majority_count(data):
    plt.figure(figsize=(12,6))
    g = sb.countplot(data.Sex)
    plt.title('MAJORITY PASSENGER COUNTS', weight='bold')
    plt.grid(linestyle = '--', axis='y')
    for i in ['top','left','right']:
        g.spines[i].set_visible(False)
    plt.tight_layout()
    
    
visual_majority_count(A)
```

```python
# Men survival rate
men = A[A.Sex=='male']['Survived']
men_rate = sum(men)/len(men)
print('The survival rate of male candidates in titanic was: {}%'.format(round(men_rate*100,2)))
```

```python
# Women survival rate
women = A[A.Sex == 'female']['Survived']
women_rate = sum(women)/len(women)
print('The survival rate of female candidates in titanic was: {}%'.format(round(women_rate * 100,2)))
```

```python
def find_survival_rate(A):
    for val in ['Pclass','SibSp','Parch','Sex']:
        plt.figure(figsize=(16,6))
        g = sb.countplot(A[val], hue = A.Survived)
        plt.title(val)
        plt.grid(linestyle = '--', axis='y')

        for i in ['top','left','right']:
            g.spines[i].set_visible(False)
        plt.show()
        
find_survival_rate(A)
```
